# Token Count

```python

# ======================================================
# IMPORTS
# ======================================================

import os
from collections import defaultdict
from mptf_parser import Corpus


# ======================================================
# CONFIG
# ======================================================

input_folder = r"C:\Users\rahaa\Dropbox\MPCD\app_exports-28-1-2026"
output_folder = r"C:\Users\rahaa\Dropbox\MPCD\processed_conllu"


# ======================================================
# LOAD CORPUS (CSV → CONLLU → SENTENCES)
# ======================================================

print("\n--- Loading corpus via MPTF parser ---")

corpus = Corpus(input_folder, output_folder)
sentences = corpus.get_sentences()

print("\n✔ Corpus loaded successfully.")
print(f"Total sentences: {len(sentences)}")


# ======================================================
# TOKEN COUNTS PER FILE
# ======================================================

tokens_per_file = defaultdict(int)

for sentence in sentences:
    fname = sentence.file_name   # ← comes from your parser
    tokens_per_file[fname] += len(sentence.get_tokens())

print("\n📄 Token counts per file:")
for fname in sorted(tokens_per_file):
    print(f"  {fname}: {tokens_per_file[fname]}")


# ======================================================
# TOTAL TOKEN COUNT
# ======================================================

total_tokens = sum(tokens_per_file.values())

print("\n🔢 Total token count:")
print(f"  {total_tokens}")


# ======================================================
# SENTENCE + TOKEN STATS PER FILE
# ======================================================

stats_per_file = defaultdict(lambda: {"sentences": 0, "tokens": 0})

for sentence in sentences:
    fname = sentence.file_name
    stats_per_file[fname]["sentences"] += 1
    stats_per_file[fname]["tokens"] += len(sentence.get_tokens())

print("\n📊 File statistics:")
for fname in sorted(stats_per_file):
    stats = stats_per_file[fname]
    print(f"  {fname}: {stats['sentences']} sentences, {stats['tokens']} tokens")


# ======================================================
# UNIQUE FILE COUNT
# ======================================================

unique_files = len(stats_per_file)

print("\n📁 Number of source files:")
print(f"  {unique_files}")
```
